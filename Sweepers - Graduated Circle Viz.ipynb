```python
import pandas as pd
import sys
import os

from mapboxgl.viz import *
from mapboxgl.utils import *
```

```python
token = os.getenv('MAPBOX_ACCESS_TOKEN')
```

```python
df = pd.read_csv('sweeper_coordinate_test.csv', nrows=9)
df
```

```python
df_to_geojson(df, filename ='../DSF Street Sweeping Project/citations.geojson', 
              properties = ['Weekday','Issue Time', 'Ticket Number', 'Route'], 
              lat = 'Latitude', lon = 'Longitude',
              precision = 3)
```

```python
# Radius stops for linear interpolation
measure = 'Issue Time'
color_breaks = [round(df[measure].quantile(q=x*0.1), 2) for x in range(1,9)]
color_stops = create_color_stops(color_breaks, colors='YlOrRd')

measure_radius = 'Weekday'
radius_breaks = [round(df[measure_radius].quantile(q=x*0.1), 2) for x in range(1,10)]
radius_stops = create_radius_stops(radius_breaks, 1, 10)


# Initialize Graduated Circle Visualization 
viz2 = GraduatedCircleViz('../DSF Street Sweeping Project/citations.geojson', 
                          access_token=token,
                          color_property = "Issue Time",
                          color_stops = color_stops,
                          radius_property = "Weekday",
                          radius_stops = radius_stops,
                          stroke_color = 'black',
                          stroke_width = 0.5,
                          center = (-118.2,34),
                          zoom = 9,
                          opacity=0.75,
                          below_layer = 'waterway-label')
viz2.label_property = "Weekday"
viz2.stroke_width = 0
viz2.label_size = 8
viz2.style='mapbox://styles/mapbox/dark-v9?optimize=true'
viz2.label_color = 'hsl(0, 0%, 70%)'
viz2.label_halo_color = 'hsla(0, 0%, 10%, 0.75)'
viz2.show()
```

This visualization alters radius size according to the day of the week, with Saturday being the largest and Sunday being the smallest. It also displays different colors for the different Issue Times, with darker colors representing later Issue Times.
