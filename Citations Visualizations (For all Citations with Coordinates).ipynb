```python
import pandas as pd
import sys
import os

from mapboxgl.viz import *
from mapboxgl.utils import *
```

```python
token = os.getenv('MAPBOX_ACCESS_TOKEN')
```

```python
#the following visualizations are for all available coordinates
#still missing about 1/3 of coordinates

df = pd.read_csv('sweeps_parking_citations.csv')
```

```python
df_to_geojson(df, filename ='../DSF Street Sweeping Project/sweeps_parking_citations.geojson', 
              properties = ['Weekday','Issue Time', 'Ticket Number', 'Route'], 
              lat = 'lat', lon = 'long',
              precision = 3)
```

```python
viz1 = CircleViz('../DSF Street Sweeping Project/sweeps_parking_citations.geojson', 
                 access_token=token, color_default='red' , 
                radius = 0.5, center = (-118, 34), zoom = 9)
viz1.show()
```

```python
with open ('viz1.html', 'w') as f:
    f.write(viz1.create_html())
```

```python
measure = 'Issue Time'
color_breaks = [round(df[measure].quantile(q=x*0.1), 2) for x in range(1,10)]
color_stops = create_color_stops(color_breaks, colors='YlOrRd')


viz2 = CircleViz('../DSF Street Sweeping Project/sweeps_parking_citations.geojson', access_token=token, 
                          center = (-118, 34), zoom = 9,
                          color_property = "Issue Time",
                          color_stops = color_stops,
                          
                          stroke_color = 'black',
                          stroke_width = 0.5,
                          opacity=0.75,
                          below_layer = 'waterway-label')
viz2.show()
```

```python
radius_stops = [
    [0.0, 0.5],
    [1.0, 1.25],
    [2.0, 2.0],
    [3.0, 2.75],
    [4.0, 3.5],
    [5.0, 4.25],
    [6.0, 5],
]


# Create the viz
viz3 = GraduatedCircleViz('../DSF Street Sweeping Project/sweeps_parking_citations.geojson',
                          access_token=token,
                          #color_property = "Issue Time",
                          #color_stops = color_stops,
                          color_default = 'red',
                          opacity = 0.6,
                          radius_property = "Weekday",
                          stroke_color = 'black',
                          stroke_width = 0.5,
                          radius_stops = radius_stops,
                          center = (-118.4, 34.1),
                          zoom = 9,
                          below_layer = 'waterway-label',
                        )

viz3.show()
```

```python
# Radius stops for linear interpolation
measure_color = 'Issue Time'
color_breaks = [round(df[measure_color].quantile(q=x*0.1), 2) for x in range(2, 10)]
color_stops = create_color_stops(color_breaks, colors='Blues')

# Generate radius breaks from data domain and circle-radius range
radius_stops = [
    [0.0, 0.5],
    [1.0, 1.25],
    [2.0, 2.0],
    [3.0, 2.75],
    [4.0, 3.5],
    [5.0, 4.25],
    [6.0, 5],
]


# Create the viz
viz4 = GraduatedCircleViz('../DSF Street Sweeping Project/sweeps_parking_citations.geojson',
                          access_token=token,
                          color_property = "Issue Time",
                          color_stops = color_stops,
                          radius_property = "Weekday",
                          stroke_color = 'black',
                          stroke_width = 0.5,
                          radius_stops = radius_stops,
                          center = (-118.2, 34),
                          zoom = 9,
                          below_layer = 'waterway-label'
                        )

viz4.show()
```
